```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '/content/drive/MyDrive/Dataset/extended_marketing_data.xlsx'
```

```python
df = pd.read_excel(path)
df.head(15)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Total Spent', hue='Comments', data=df)
plt.title('Total Spent vs Comments')
plt.xlabel('Total Spent')
plt.ylabel('Comments')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Age', hue='Total Spent', data=df)
plt.title('Age vs Total Spent')
plt.xlabel('Age')
plt.ylabel('Total Spent')
plt.show()
```

```python
path = '/content/drive/MyDrive/Dataset/customer_data.xlsx'
```

```python
df = pd.read_excel(path)
df.head(15)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Total Spent ($)', hue='Gender', data=df)
plt.title('Gender vs Total Spent')
plt.xlabel('Total Spent ($)')
plt.ylabel('Gender')
plt.show()
print(df.columns)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Total Spent ($)', hue='Gender', data=df)
plt.show()
```

```python
df= pd.read_excel(path)
sns.violinplot(x='Total Spent ($)', y='Age', data=df, palette='Set2')
plt.title('Total Spent vs Age')
plt.xlabel('Total Spent ($)')
plt.ylabel('Age')
plt.show()
```
