```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/Dataset/complex_marketing_data_enriched.xlsx'
```

```python
df = pd.read_excel(path)
```

```python
df['Total Cost'] = df['Total Spent']
df['Profitability'] = df['Total Cost'] - df['Stock']
df['Profitability (%)'] = (df['Profitability'] / df['Stock'])

print(df[['Customer ID', 'Total Cost', 'Profitability', 'Profitability (%)']])

df.plot(
    x='Customer ID',
    y='Profitability',
    kind='bar',
    title='Customer Profitability Analysis'
)
```

```python
df['Stock'] = pd.to_numeric(df['Stock'], errors='coerce')
df['Cost per Stock'] = df['Total Spent'] / df['Stock']

print(df[['Customer ID', 'Total Spent', 'Stock', 'Cost per Stock']])
```

```python
df.plot(
    x='Discount Availed',
    y='Total Spent',
    kind='scatter',
    title='Discount vs Total Spending'

)
```

```python
region_spending = df.groupby('Region')['Total Spent'].mean().reset_index()
region_spending.plot(
    x='Region',
    y='Total Spent',
    kind='bar',
    title='Region-Wise Average Spending'
)
```

```python
import pandas as pd

activity_data = {
    'Activity': ['Order Processing', 'Customer Support', 'Marketing Campaign'],
    'Total Cost': [10000, 5000, 8000],
    'Cost Driver': ['Number of Orders', 'Support Tickets', 'Marketing Campaigns'],
    'Driver Volume': [200, 50, 10]
}

activity_df = pd.DataFrame(activity_data)
activity_df['Cost per Driver Unit'] = activity_df['Total Cost'] / activity_df['Driver Volume']


customer_activity_data = {
    'Customer ID': [1, 2, 3, 4],
    'Number of Orders': [10, 30, 20, 40],
    'Support Tickets': [2, 5, 1, 3],
    'Marketing Campaigns': [1, 2, 1, 1]
}

customer_df = pd.DataFrame(customer_activity_data)
customer_df['Order Processing Cost'] = customer_df['Number of Orders'] * activity_df.loc[0, 'Cost per Driver Unit']
customer_df['Support Cost'] = customer_df['Support Tickets'] * activity_df.loc[1, 'Cost per Driver Unit']
customer_df['Marketing Cost'] = customer_df['Marketing Campaigns'] * activity_df.loc[2, 'Cost per Driver Unit']

customer_df['Total Allocated Cost'] = (
    customer_df['Order Processing Cost'] +
    customer_df['Support Cost'] +
    customer_df['Marketing Cost']
)

print(customer_df)

customer_df.plot(
    x='Customer ID',
    y=['Order Processing Cost', 'Support Cost', 'Marketing Cost'],
    kind='bar',
    stacked=True,
    title='Activity-Based Costing'
)
```
