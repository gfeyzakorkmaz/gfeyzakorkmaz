```python
def calculate_sales_growth_rate():
  sales_df["Growth Rate (%)"] = sales_df["Total Sales ($)"].pct_change() * 100
  sales_df["Growth Rate (%)"].plot(kind="bar", figsize=(10, 6))
  plt.title("Annual Sales Growth Rate")
  plt.ylabel("Growth Rate (%)")
  plt.xlabel("Year")
  plt.grid()
  plt.show()

calculate_sales_growth_rate()
```

```python
def department_budget_share():
  total_budget_by_department = department_df.groupby("Department")["Budget ($)"].sum()
  total_budget_by_department.plot(kind="pie", autopct="1.1f%%", figsize=(8, 8))
  plt.title("Share of Total Budget by Department")
  plt.ylabel("")
  plt.show()

department_budget_share()
```

```python
def marketing_expense_change_analysis():
  marketing_budget_df["Yearly Change (%)"] = marketing_budget_df["Total Marketing Expense ($)"].pct_change() * 100
  marketing_budget_df["Yearly Change (%)"].plot(kind="line", figsize=(10, 6), marker="o")
  plt.title("Annual Change in Marketing Expenses")
  plt.ylabel("Change (%)")
  plt.xlabel("Year")
  plt.grid()
  plt.show()

marketing_expense_change_analysis()
```

```python
def product_sales_share():
  product_sales = sales_df[["Laptops", "Desktops", "Accessories"]].sum()
  product_sales.plot(kind="pie", autopct="%1.1f%%", figsize=(8, 8))
  plt.title("Contribution of Product Groups to Total Sales")
  plt.ylabel("")
  plt.show()

product_sales_share()
```
