```python
import pandas as pd
df = pd.read_excel(path)
df.head(15)
```

```python
X = df[['Age', 'Total Spent']]
k=3
kmeans = KMeans(n_clusters=k, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

plt.scatter(df['Age'],
  df['Total Spent'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Total Spent')
plt.title('Customer Sementation with K-means')
plt.show()
```
