```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '/content/drive/MyDrive/Dataset/extended_marketing_data.xlsx'
```

```python
df = pd.read_excel(path)
```

```python
df['Gender'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Total Spent', hue='Comments', data=df)
plt.title('Total Spent vs Comments')
plt.xlabel('Total Spent')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Gender', hue='Total Spent', data=df)
plt.title('Total Spent vs Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Age', hue='Total Spent', data=df)
plt.title('Total Spent vs Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Age', y='Total Spent', data=df)
plt.title('Total Spent vs Age')
plt.xlabel('Age')
plt.ylabel('Total Spent')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
pivot_table = df.pivot_table(index='Age', columns='Gender', values='Total Spent', aggfunc='mean')
sns.heatmap(pivot_table, annot=True, cmap='coolwarm')
plt.title('Heatmap of Total Spent by Age and Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10,6))
pivot_table = df.pivot_table(index='Comments', columns='Age', values='Total Spent', aggfunc='mean')
sns.heatmap(pivot_table, annot=True, cmap='coolwarm')
plt.title('Heatmap of Total Spent by Age and Comments')
plt.xlabel('Age')
plt.ylabel('Comments')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='Age', y='Total Spent', hue='Gender', data=df, split=True)
plt.title('Total Spent vs Age and Gender')
plt.xlabel('Age')
plt.ylabel('Total Spent')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x='Age', y='Total Spent', hue='Gender', data=df)
plt.title('Total Spent vs Age and Gender')
plt.xlabel('Age')
plt.ylabel('Total Spent')
plt.xticks(rotation=45)
plt.show()
```
