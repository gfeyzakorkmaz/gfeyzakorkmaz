```python
import pandas as pd
```

```python
data = {
    'Customer ID': [1, 2, 3, 4],
    'Revenue': [1000, 1500, 800, 2000],
    'Cost of Goods Sold': [400, 600, 300, 700],
    'Marketing Cost': [100, 200, 100, 300],
    'Support Cost': [50, 100, 50, 150]
}

df = pd.DataFrame(data)

df['Total Cost'] = df['Cost of Goods Sold'] + df['Marketing Cost'] + df['Support Cost']
df['Profitability'] = df['Revenue'] - df['Total Cost']

df['Profitability (%)'] = (df['Profitability'] / df['Revenue']) * 100

print(df)

df.plot(x='Customer ID', y='Profitability', kind='bar', title='Customer Profitability Analysis')
```

Calculate customer profitability analysis (CPA)

```python

activity_data = {
    'Activity': ['Order Processing', 'Customer Support', 'Marketing Campaign'],
    'Total Cost': [10000, 5000, 8000],
    'Cost Driver': ['Number of Orders', 'Support Tickets', 'Marketing Campaigns'],
    'Driver Volume': [200, 50, 10]
}

activity_df = pd.DataFrame(activity_data)


activity_df['Cost per Driver Unit'] = activity_df['Total Cost'] / activity_df['Driver Volume']


customer_activity_data = {
    'Customer ID': [1, 2, 3, 4],
    'Number of Orders': [10, 30, 20, 40],
    'Support Tickets': [2, 5, 1, 3],
    'Marketing Campaigns': [1, 2, 1, 1]
}

customer_df = pd.DataFrame(customer_activity_data)


customer_df['Order Processing Cost'] = customer_df['Number of Orders'] * activity_df.loc[0, 'Cost per Driver Unit']
customer_df['Support Cost'] = customer_df['Support Tickets'] * activity_df.loc[1, 'Cost per Driver Unit']
customer_df['Marketing Cost'] = customer_df['Marketing Campaigns'] * activity_df.loc[2, 'Cost per Driver Unit']


customer_df['Total Allocated Cost'] = (
    customer_df['Order Processing Cost'] +
    customer_df['Support Cost'] +
    customer_df['Marketing Cost']
)


print(customer_df)


customer_df.plot(x='Customer ID', y=['Order Processing Cost', 'Support Cost', 'Marketing Cost'], kind='bar', stacked=True, title='Activity-Based Costing')
```

Activity-Based Costing (ABC)

Customer Profitability Analysis (CPA):

Calculates profitability per customer using revenue and costs (e.g., marketing, support).
Outputs profit values and their percentage contribution.
Activity-Based Costing (ABC):

Allocates costs based on activity drivers (e.g., number of orders, support tickets).
Computes individual and total activity costs for each custome (from ChatGPT)
