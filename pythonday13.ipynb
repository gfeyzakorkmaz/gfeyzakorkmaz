```python
import pandas as pd
```

```python
from google.colab import data_table
from vega_datasets import data
import matplotlib.pyplot as plt
```

```python
df = data.gapminder()
print(df.head())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
countries = ['United States', 'China', 'India', 'Afghanistan']
df_filtered = df[df['country'].isin(countries)]

for country in countries:
  subset = df_filtered[df_filtered['country'] == country]
  plt.plot(subset['year'], subset['life_expect'], label=country)

  plt.title('Life Expectancy Over Time')
  plt.xlabel('Year')
  plt.ylabel('Life Expectancy')
  plt.legend()
  plt.show()
```

```python
df_1980 = df[df['year'] == 1980]

df_1980.nlargest(10, 'gdpPercap').plot(kind='bar', x='country', y='gdpPercap', color='purple')
plt.title('Top 10 Countries by GDP in 2007')
plt.ylabel('GDP per Capita')
plt.xlabel('Country')
plt.show()

print(df_1980.columns)
```

```python
df_2007 = df[df['year'] == 2007]


df_2007.nlargest(10, 'gdpPercap').plot(kind='bar', x='country', y='gdpPercap', color='purple')

plt.title('Top 10 Countries by GDP in 2007')
plt.ylabel('GDP per Capita')
plt.xlabel('Country')
plt.show()

print(df_2007.columns)
```

```python
df_2007 = df[df['year'] == 2007]
top_10_gdp = df_2007.nlargest(10, 'gdpPercap')

top_10_gdp.plot(kind='bar', x='country', y='gdpPercap', color='purple')

plt.title('Top 10 Countries by GDP per Capita in 1980')
plt.ylabel('GDP per Capita')
plt.xlabel('Country')
plt.show()

print(df_2007.columns)
```
