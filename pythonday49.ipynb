```python
import pandas as pd
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis
from gensim import corpora
from gensim.models import LdaModel

df = pd.read_excel(path)

documents = df['Comments'].tolist()
texts = [[word for word in document.lower().split()] for document in documents]
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]


lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=10, random_state=42)

lda_vis = gensimvis.prepare(lda_model, corpus, dictionary)

pyLDAvis.save_html(lda_vis, 'lda_visualization.html')
pyLDAvis.display(lda_vis)
```

```python
for idx, topic in lda_model.show_topics(formatted=False, num_words=10):
   print(f"Topic {idx + 1}: {[word for word, _ in topic]}")
```

```python
doc_topics = [lda_model.get_document_topics(bow) for bow in corpus]
for i, topics in enumerate(doc_topics[:10]):
   print(f"Document {i + 1}: {topics}")
```

```python
from gensim.corpora.dictionary import Dictionary

comments = df['Comments'].dropna().str.lower().str.split()
dictionary = Dictionary(comments)
corpus = [dictionary.doc2bow(text) for text in comments]

lda = LdaModel(corpus=corpus, num_topics=5, id2word=dictionary, random_state=42)

lda_display = gensimvis.prepare(lda, corpus, dictionary)
pyLDAvis.display(lda_display)
```
